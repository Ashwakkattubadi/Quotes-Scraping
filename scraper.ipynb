```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page=1
while True:
    url = f"https://quotes.toscrape.com/page/{page}/"
    res=requests.get(url)
    soup = BeautifulSoup(res.text, "lxml")
    quotes=soup.find_all("div", class_="quote")

    if not quotes:
        print("No more pages...")
        break
        
    with open(f"scraped_data/page{page}.html","w", encoding="utf-8") as f:
        f.write(res.text)
        print(f"Downloaded data from page{page}")

    page+=1

last_page=page
```

```python
data=[]
for page in range(1, last_page):
    with open(f"scraped_data/page{page}.html", "r", encoding="utf-8") as f:
        html_content=f.read()
    soup=BeautifulSoup(html_content, "lxml")
    quotes_blocks=soup.select("div.quote")

    for block in quotes_blocks:
      quote=block.select_one("span.text").get_text(strip=True)
      author=block.find("small", class_="author").get_text(strip=True)
      data.append({"Quote":quote,
                "Author":author})   
```

```python
df=pd.DataFrame(data)
df.head(20)
```
